## FASE 2: Cargar y limpiar los DataFrames

En esta celda se cargan los CSV sucios y se limpian los datos en memoria.

```python
import pandas as pd

productos = pd.read_csv(WORKDIR / "parcial_2_productos_sucios.csv")
clientes = pd.read_csv(WORKDIR / "parcial_2_clientes_sucios.csv")
proveedores = pd.read_csv(WORKDIR / "parcial_2_proveedores_sucios.csv")

productos[["producto", "categoria"]] = (
    productos["producto_categoria"].str.split(" - ", n=1, expand=True)
)
productos["producto"] = productos["producto"].str.strip().str.title()
productos["categoria"] = productos["categoria"].str.strip().str.title()
productos["precio"] = (
    productos["precio"].astype(str)
    .str.replace("$", "", regex=False)
    .str.replace(",", "", regex=False)
)
productos["precio"] = pd.to_numeric(productos["precio"], errors="coerce").fillna(0).astype(int)
productos["stock"] = pd.to_numeric(productos["stock"], errors="coerce").fillna(0).astype(int)
productos["fecha_vencimiento"] = pd.to_datetime(productos["fecha_vencimiento"], errors="coerce").dt.strftime("%Y-%m-%d")
productos = productos[["id_producto", "producto", "categoria", "precio", "stock", "fecha_vencimiento"]]

clientes[["nombre", "tipo"]] = (
    clientes["cliente_tipo"].astype(str).str.split(" - ", n=1, expand=True)
)
clientes["nombre"] = clientes["nombre"].str.strip().str.title()
clientes["tipo"] = clientes["tipo"].str.strip().str.title()
clientes["email"] = clientes["email"].fillna("No Registra")
clientes["telefono"] = clientes["telefono"].replace({"NaN": pd.NA}).fillna("No Registra")
clientes["fecha_nacimiento"] = pd.to_datetime(clientes["fecha_nacimiento"], errors="coerce").dt.strftime("%Y-%m-%d")
clientes = clientes[["id_cliente", "nombre", "tipo", "email", "telefono", "fecha_nacimiento"]]

proveedores[["empresa", "ciudad"]] = (
    proveedores["empresa_ciudad"].astype(str).str.split(" - ", n=1, expand=True)
)
proveedores["empresa"] = proveedores["empresa"].str.strip().str.title()
proveedores["ciudad"] = proveedores["ciudad"].str.strip().str.title()
proveedores["contacto"] = proveedores["contacto"].replace({"NaN": pd.NA}).fillna("No Registra").str.title()
proveedores["telefono"] = proveedores["telefono"].replace({"NaN": pd.NA}).fillna("No Registra")
proveedores["email"] = proveedores["email"].fillna("No Registra")
proveedores = proveedores[["nit_proveedor", "empresa", "ciudad", "contacto", "telefono", "email"]]

print("DataFrames limpiados")
print(productos.head(3))
print(clientes.head(3))
print(proveedores.head(3))
```

## FASE 3: Migración a SQLite

Crear la base de datos `parcial_2_cafeteria.db` y cargar las tablas de productos, clientes y proveedores.

```python
import sqlite3

DB_PATH = WORKDIR / "parcial_2_cafeteria.db"
connection = sqlite3.connect(DB_PATH)
connection.execute("PRAGMA foreign_keys = ON;")

with connection:
    connection.execute("DROP TABLE IF EXISTS ventas")
    connection.execute("DROP TABLE IF EXISTS productos")
    connection.execute("DROP TABLE IF EXISTS clientes")
    connection.execute("DROP TABLE IF EXISTS proveedores")

    connection.execute("""
        CREATE TABLE productos (
            id_producto INTEGER PRIMARY KEY,
            producto TEXT NOT NULL,
            categoria TEXT NOT NULL,
            precio INTEGER NOT NULL,
            stock INTEGER NOT NULL,
            fecha_vencimiento TEXT
        )
        """)
    connection.execute("""
        CREATE TABLE clientes (
            id_cliente INTEGER PRIMARY KEY,
            nombre TEXT NOT NULL,
            tipo TEXT NOT NULL,
            email TEXT NOT NULL,
            telefono TEXT NOT NULL,
            fecha_nacimiento TEXT NOT NULL
        )
        """)
    connection.execute("""
        CREATE TABLE proveedores (
            nit_proveedor INTEGER PRIMARY KEY,
            empresa TEXT NOT NULL,
            ciudad TEXT NOT NULL,
            contacto TEXT NOT NULL,
            telefono TEXT NOT NULL,
            email TEXT NOT NULL
        )
        """)

    connection.executemany("INSERT INTO productos (id_producto, producto, categoria, precio, stock, fecha_vencimiento) VALUES (?, ?, ?, ?, ?, ?)",
        productos.values.tolist(),
    )
    connection.executemany("INSERT INTO clientes (id_cliente, nombre, tipo, email, telefono, fecha_nacimiento) VALUES (?, ?, ?, ?, ?, ?)",
        clientes.values.tolist(),
    )
    connection.executemany("INSERT INTO proveedores (nit_proveedor, empresa, ciudad, contacto, telefono, email) VALUES (?, ?, ?, ?, ?, ?)",
        proveedores.values.tolist(),
    )

print("SQLite: tablas cargadas en", DB_PATH)
```

## FASE 4: Crear la tabla `ventas` y ejecutar CRUD

Crear la tabla de ventas con llaves foráneas y ejecutar los comandos SQL solicitados.

```python
with connection:
    connection.execute("PRAGMA foreign_keys = ON;")
    connection.execute("""
        CREATE TABLE IF NOT EXISTS ventas (
            id_venta INTEGER PRIMARY KEY AUTOINCREMENT,
            id_cliente INTEGER NOT NULL,
            id_producto INTEGER NOT NULL,
            id_proveedor INTEGER NOT NULL,
            cantidad INTEGER NOT NULL,
            total_venta REAL NOT NULL,
            fecha_venta TEXT NOT NULL,
            FOREIGN KEY (id_cliente) REFERENCES clientes(id_cliente),
            FOREIGN KEY (id_producto) REFERENCES productos(id_producto),
            FOREIGN KEY (id_proveedor) REFERENCES proveedores(nit_proveedor)
        )
        """)

    ventas = [
        (101, 1, 900111, 2, 10000.0, "2026-05-10"),
        (102, 3, 700333, 1, 3000.0, "2026-05-11"),
        (103, 5, 600444, 3, 7500.0, "2026-05-12"),
        (104, 8, 100999, 4, 14000.0, "2026-05-13"),
        (105, 20, 200888, 1, 2000.0, "2026-05-14"),
    ]

    connection.executemany("INSERT INTO ventas (id_cliente, id_producto, id_proveedor, cantidad, total_venta, fecha_venta) VALUES (?, ?, ?, ?, ?, ?)",
        ventas,
    )

    resultado = pd.read_sql_query("""
        SELECT v.id_venta, c.nombre AS cliente, p.producto AS producto, v.total_venta
        FROM ventas v
        INNER JOIN clientes c ON v.id_cliente = c.id_cliente
        INNER JOIN productos p ON v.id_producto = p.id_producto
        """,
        connection,
    )
    print("Resultados SELECT:")
    print(resultado)

    connection.execute("UPDATE ventas SET cantidad = 5, total_venta = 25000.0 WHERE id_venta = 1")
    connection.execute("DELETE FROM ventas WHERE id_venta = 3")

ventas_final = pd.read_sql_query("SELECT * FROM ventas ORDER BY id_venta", connection)
print("Ventas finales:")
print(ventas_final)

connection.close()
```
